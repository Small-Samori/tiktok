```python
import os, glob
import pandas as pd
from pandas.errors import EmptyDataError
from tqdm import tqdm
```

```python
oak = os.getenv('OAK')
video_folder = f"{oak}/samori/tiktok/videos"
comments_folder = f"{oak}/samori/tiktok/comments"
os.path.exists(comments_folder)
```

```python
months = os.listdir(video_folder)
months
```

```python
for month in months:
    # print(f"---- {month} ----")
    month_folder = f"{video_folder}/{month}"
    csvs = glob.glob(f"{month_folder}/*.csv")
    month_combined_df = pd.read_csv(csvs[0])
    for path in tqdm(csvs[1:]):
        try:  
            df = pd.read_csv(path)
        except EmptyDataError:
            continue
        month_combined_df = pd.concat([month_combined_df, df], ignore_index=True)

    month_combined_df = month_combined_df.drop_duplicates(subset='id', keep='first')
    month_combined_df.to_csv(f"{video_folder}/combined/{month}.csv", index=False)
    print(f"\n{month}: {len(month_combined_df)} recordz ...")
    print('\n')
```

```python
combined_csvs = glob.glob(f"{video_folder}/combined/*.csv")
all_combined = pd.read_csv(combined_csvs[0])
for path in combined_csvs[1:]:
    df = pd.read_csv(path)
    all_combined = pd.concat([all_combined, df], ignore_index=True)
all_combined.to_csv(f"{video_folder}/combined/all_months.csv", index=False)
```
