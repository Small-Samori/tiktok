```python
from utils_lda import *
import pickle, os
import gensim
import pandas as pd
import numpy as np
```

```python
data_name = "rd_tt_combined_2"

data_dropped = pd.read_csv(f"./{data_name}/df_dropped.csv")

with open(f"./{data_name}/lemmatized_data_text_dropped.pickle", "rb") as f:
    lemmatized_data_text_dropped = pickle.load(f)

with open(f"./{data_name}/id2word.pickle", "rb") as f:
    id2word = pickle.load(f)

with open(f"./{data_name}/corpus.pickle", "rb") as f:
    corpus = pickle.load(f)

home = os.getenv("HOME")
os.environ.update({'MALLET_HOME':r'/home/users/iasamori/tiktok/mallet-2.0.8/'})
mallet_path = f"{home}/tiktok/mallet-2.0.8/bin/mallet"
```

```python
topic_distribution = model[corpus]
```

```python
all_topics_scores = []
for post_num in range(len(topic_distribution)):
    post = topic_distribution[post_num]
    topic_scores = []

    for i in range(num_topics):
        topic_scores.append(post[i][1])
        
    all_topics_scores.append(topic_scores)
```

```python
def rank_biserial_correlation(x, y):
    n_x = len(x)
    n_y = len(y)
    # Use Mann-Whitney U test. By default, mannwhitneyu returns the smaller U value.
    U_stat, p = stats.mannwhitneyu(x, y, alternative='two-sided')
    # Calculate rank biserial correlation.
    # Note: This formulation assumes that a higher score in x tends to indicate a different group ordering.
    r = 1 - (2 * U_stat) / (n_x * n_y)
    return r
```
